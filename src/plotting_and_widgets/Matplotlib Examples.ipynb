````python
%matplotlib notebook
import matplotlib.pyplot as plt

print(f'interactive mode is: {plt.isinteractive()}')

input_values = range(1,1001)
squares = [i**2 for i in input_values]
cubes = [i**3 for i in input_values]

plt.style.use('seaborn')
fig1, ax1 = plt.subplots()

ax1.scatter(input_values, cubes, c=cubes, cmap=plt.cm.flag,   s=10)

ax1.set_title("Cube Numbers", fontsize=24)
ax1.set_xlabel("Value", fontsize=14)
ax1.set_ylabel("Cube of Value", fontsize=14)
ax1.tick_params(axis='both', which='major', labelsize=14)
ax1.set_title("Cube Numbers", fontsize=24)
````

````python
ax1.set_title("Cube Numbers Revised", fontsize=24)
````

````python
fig2, ax2 = plt.subplots()

ax2.scatter(input_values, squares, c=squares, cmap=plt.cm.flag,   s=10)

ax2.set_title("Square Numbers", fontsize=24)
ax2.set_xlabel("Value", fontsize=14)
ax2.set_ylabel("Square of Value", fontsize=14)
ax2.tick_params(axis='both', which='major', labelsize=14)
ax2.set_title("Square Numbers", fontsize=24)
````

````python
ax1.set_title("Cube #s", fontsize=24)
ax2.set_title("Square #s", fontsize=24)
````

````python
import numpy as np

x = np.linspace(0, 2, 100)

# Note that even in the OO-style, we use `.pyplot.figure` to create the figure.
fig, ax = plt.subplots()  # Create a figure and an axes.
ax.plot(x, x, label='linear')  # Plot some data on the axes.
ax.plot(x, x**2, label='quadratic')  # Plot more data on the axes...
ax.plot(x, x**3, label='cubic')  # ... and some more.
ax.set_xlabel('x label')  # Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.
ax.set_title("Simple Plot")  # Add a title to the axes.
ax.legend()  # Add a legend.
````
